```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dir_img = '180119'
img_name = 'SAM_6390.JPG'
img = cv2.imread(dir_img + '\\' + img_name, 0)  # convert to grayscale
plt.imshow(img, 'gray')
plt.show()
```

```python
import json
from pprint import pprint

with open('via_region_data.json') as f:
    data = json.load(f)

pprint(data)
```

```python
# extract shape_attributes values (from json file)
shape_att = data[img_name + '425953']['regions'][0]['shape_attributes']
shape_att
```

```python
import matplotlib.patches as ptc

# adding annotation to the image
fig, ax = plt.subplots(1)
ax.imshow(img, 'gray')
rect = ptc.Rectangle((shape_att['x'], shape_att['y']),
                     shape_att['width'], shape_att['height'],
                     linewidth=1, edgecolor='r', facecolor='none')
ax.add_patch(rect)
plt.show()
```

```python
import json
from pprint import pprint

with open('via_region_data_2.json') as f:
    data = json.load(f)

pprint(data)
```

```python
# extract shape_attributes values (from json file)
shape_att = data[img_name + '425953']['regions']
shape_att_i = []

for i in range(len(shape_att)):
    shape_att_i.append(shape_att[i]['shape_attributes'])
    
```

```python
import matplotlib.patches as ptc

# adding annotation to the image
fig, ax = plt.subplots(1)
ax.imshow(img, 'gray')

x = []; y = []; w = []; h = []

for i in range(len(shape_att_i)):
    x.append(shape_att_i[i]['x'])
    y.append(shape_att_i[i]['y'])
    w.append(shape_att_i[i]['width'])
    h.append(shape_att_i[i]['height'])
    rect = ptc.Rectangle((x[i], y[i]), w[i], h[i], linewidth=1,
                         edgecolor='r', facecolor='none')
    ax.add_patch(rect)
    
plt.show()
```
